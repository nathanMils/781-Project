# 781 Project

## having_IP_Address {-1, 1}
- **Description**: Indicates if the domain part has an IP address.
- **Rule**:
  - If the domain part has an IP address → **Phishing** (`-1`)
  - Otherwise → **Legitimate** (`1`)

## URL_Length {1, 0, -1}
- **Description**: Categorizes URLs based on their length.
- **Rule**:
  - If URL length < 54 → **Legitimate** (`1`)
  - If URL length ≥ 54 and ≤ 75 → **Suspicious** (`0`)
  - Otherwise → **Phishing** (`-1`)

## Shortening_Service {1, -1}
- **Description**: Determines if a URL shortening service was used.
- **Rule**:
  - If using TinyURL or similar services → **Phishing** (`-1`)
  - Otherwise → **Legitimate** (`1`)

## having_At_Symbol {1, -1}
- **Description**: Checks if the URL contains an "@" symbol.
- **Rule**:
  - If URL contains "@" symbol → **Phishing** (`-1`)
  - Otherwise → **Legitimate** (`1`)

## double_slash_redirecting {-1, 1}
- **Description**: Evaluates the presence of "//" in the URL.
- **Rule**:
  - If "//" appears in the URL after position 7 → **Phishing** (`-1`)
  - Otherwise → **Legitimate** (`1`)

## Prefix_Suffix {-1, 1}
- **Description**: Checks if URL starts or ends with hyphens or dots.
- **Rule**:
  - If URL has hyphens/dots at start or end → **Phishing** (`-1`)
  - Otherwise → **Legitimate** (`1`)

## having_Sub_Domain {-1, 0, 1}
- **Description**: Assesses the number of subdomains in the URL.
- **Rule**:
  - More than one subdomain → **Phishing** (`-1`)
  - Single subdomain → **Suspicious** (`0`)
  - Otherwise → **Legitimate** (`1`)

## SSLfinal_State {-1, 1, 0}
- **Description**: Checks the validity of the SSL certificate.
- **Rule**:
  - Invalid or missing SSL certificate → **Phishing** (`-1`)
  - Valid but not fully secure SSL → **Suspicious** (`0`)
  - Otherwise → **Legitimate** (`1`)

## Domain_registration_length {-1, 1}
- **Description**: Indicates the duration of domain registration.
- **Rule**:
  - If domain registered for less than 6 months → **Phishing** (`-1`)
  - Otherwise → **Legitimate** (`1`)

## Favicon {1, -1}
- **Description**: Examines the legitimacy of the favicon.
- **Rule**:
  - If favicon exists and is legitimate → **Legitimate** (`1`)
  - Otherwise → **Phishing** (`-1`)

## port {1, -1}
- **Description**: Checks if standard ports (80, 443) are used.
- **Rule**:
  - If using standard ports → **Legitimate** (`1`)
  - Otherwise → **Phishing** (`-1`)

## HTTPS_token {-1, 1}
- **Description**: Checks if "https" is present in the URL and used correctly.
- **Rule**:
  - If "https" is absent or inconsistently used → **Phishing** (`-1`)
  - Otherwise → **Legitimate** (`1`)

## Request_URL {1, -1}
- **Description**: Verifies if URLs in the request are legitimate.
- **Rule**:
  - Legitimate URL in request → **Legitimate** (`1`)
  - Otherwise → **Phishing** (`-1`)

## URL_of_Anchor {-1, 0, 1}
- **Description**: Evaluates the nature of anchor texts in the URL.
- **Rule**:
  - Misleading or missing anchor text → **Phishing** (`-1`)
  - Generic anchor text → **Suspicious** (`0`)
  - Otherwise → **Legitimate** (`1`)

## Links_in_tags {1, -1, 0}
- **Description**: Checks the legitimacy of links within tags.
- **Rule**:
  - No links or links to legitimate sites → **Legitimate** (`1`)
  - Links to suspicious sites → **Phishing** (`-1`)
  - Otherwise → **Suspicious** (`0`)

## SFH {-1, 1, 0}
- **Description**: Assesses the form handler (SFH).
- **Rule**:
  - SFH points to a non-legitimate site → **Phishing** (`-1`)
  - SFH is a generic form handler → **Suspicious** (`0`)
  - Otherwise → **Legitimate** (`1`)

## Submitting_to_email {-1, 1}
- **Description**: Indicates if the form submits to an email address.
- **Rule**:
  - Form submission via email → **Phishing** (`-1`)
  - Otherwise → **Legitimate** (`1`)

## Abnormal_URL {-1, 1}
- **Description**: Checks for unusual characters or patterns in the URL.
- **Rule**:
  - Unusual characters or patterns → **Phishing** (`-1`)
  - Otherwise → **Legitimate** (`1`)

## Redirect {0, 1}
- **Description**: Indicates if the URL redirects to another domain.
- **Rule**:
  - Redirects to another domain → **Phishing** (`-1`)
  - No redirection → **Legitimate** (`1`)
  - Otherwise → **Suspicious** (`0`)

## on_mouseover {1, -1}
- **Description**: Verifies if suspicious behavior is triggered by mouseover.
- **Rule**:
  - Suspicious behavior on mouseover → **Phishing** (`-1`)
  - Otherwise → **Legitimate** (`1`)

## RightClick {1, -1}
- **Description**: Indicates if right-clicking is disabled on the page.
- **Rule**:
  - Right-click disabled → **Phishing** (`-1`)
  - Otherwise → **Legitimate** (`1`)

## popUpWindow {1, -1}
- **Description**: Checks for pop-ups generated without user action.
- **Rule**:
  - Unsolicited pop-ups → **Phishing** (`-1`)
  - Otherwise → **Legitimate** (`1`)

## Iframe {1, -1}
- **Description**: Indicates if iframes load content from other domains.
- **Rule**:
  - Iframes with external content → **Phishing** (`-1`)
  - Otherwise → **Legitimate** (`1`)

## age_of_domain {-1, 1}
- **Description**: Assesses the age of the domain.
- **Rule**:
  - If the domain age is below a threshold (e.g., 6 months) → **Phishing** (`-1`)
  - Otherwise → **Legitimate** (`1`)

## DNSRecord {-1, 1}
- **Description**: Checks for missing or suspicious DNS records.
- **Rule**:
  - Missing or suspicious DNS records → **Phishing** (`-1`)
  - Otherwise → **Legitimate** (`1`)

## web_traffic {-1, 0, 1}
- **Description**: Indicates the level of web traffic to the site.
- **Rule**:
  - Low or no web traffic → **Phishing** (`-1`)
  - Moderate web traffic → **Suspicious** (`0`)
  - Otherwise → **Legitimate** (`1`)

## Page_Rank {-1, 1}
- **Description**: Checks if the page rank is high.
- **Rule**:
  - Low or no page rank → **Phishing** (`-1`)
  - Otherwise → **Legitimate** (`1`)

## Google_Index {1, -1}
- **Description**: Indicates if the URL is indexed by Google.
- **Rule**:
  - Indexed by Google → **Legitimate** (`1`)
  - Otherwise → **Phishing** (`-1`)

## Links_pointing_to_page {1, 0, -1}
- **Description**: Checks the number and legitimacy of links pointing to the page.
- **Rule**:
  - Many legitimate links → **Legitimate** (`1`)
  - Few links → **Suspicious** (`0`)
  - No or bad links → **Phishing** (`-1`)

## Statistical_report {-1, 1}
- **Description**: Indicates if statistical analysis detects malicious behavior.
- **Rule**:
  - Malicious behavior → **Phishing** (`-1`)
  - Otherwise → **Legitimate** (`1`)

## Result {-1, 1}
- **Description**: Overall classification of the URL.
- **Rule**:
  - Final classification is phishing → **Phishing** (`-1`)
  - Otherwise → **Legitimate** (`1`)
